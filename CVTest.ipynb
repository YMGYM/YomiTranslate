```python
import cv2
```

```python
image = cv2.imread('TestFile/en_xs.jpg')
```

```python
ret, thes = cv2.threshold(image, 105, 255, cv2.THRESH_BINARY)
```

```python
import matplotlib.pyplot as plt
```

```python
# plt.subplots(figsize = (100,100))
plt.imshow(thes, 'gray')
```

```python
print(pyt.image_to_string(thes))
```
